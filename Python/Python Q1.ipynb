```python
# Q1
import pandas as pd

# open people_1
People1 = pd.read_table('people_1.txt')
People1.head()
# open people_2
People2 = pd.read_table('people_2.txt')
People2.head()

#combine 2 files
df= pd.concat([People1, People2])
```

```python
df.head()
```

```python
#Capitalize Firstname
df['FirstName']= df['FirstName'].str.capitalize()
#caplitalize Lastname
df['LastName']= df['LastName'].str.capitalize()
#formating phonenumber
df['Phone']= df['Phone'].str.replace('-', '')
#formating email
df['Email']= df['Email'].str.lower()
```

```python
df.head()
```

```python
# Formatting the address

df["Address"] = df["Address"].apply(lambda x: x.translate({ord(c): None for c in "No.#"}))
df
```

```python
#drop duplicates 

df = df.drop_duplicates(subset=['Address', 'FirstName', 'LastName','Email', 'Phone'])
```

```python
df.head()
```

```python
# Convert to CSV
df.to_csv ('Newfile.csv', index=None)
```
